##1. Membaca data dari url

```python
import pandas as pd
link = "https://raw.githubusercontent.com/njungger/dataset/main/data-jumlah-korban-kekerasan-berdasarkan-tempat-kejadian-perkara-bulan-desember-tahun-2020.csv"
df = pd.read_csv(link)
```

```python
df.info()
```

##2. membuat statistik diskriptif dari data (setiap atribut yang bertipe numerik)

```python
df.describe()
```

##3. mendeskripsikan kemiringan/skewned dari data setiap variabel (bertipe numerik)

```python
import seaborn as sns
sns.distplot(df['jumlah']),
hist=True,
kde=True,
bins=int(180/5),
color='red',
hist_kws={'edgecolor':'black'},
kde_kws={'linewidth':5}
```
